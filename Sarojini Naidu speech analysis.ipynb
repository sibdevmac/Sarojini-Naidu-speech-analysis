```python
import re
from collections import Counter
```

```python
clean_text = re.sub(r'[^a-zA-Z\s]', '', df.lower())
words = clean_text.split()
word_counts = Counter(words)
```

```python
nationalism_keywords = ["nation", "national", "india", "country", "renaissance", "regeneration", "freedom", "liberty"]
feminism_keywords = ["woman", "women", "womanhood", "rights", "education", "sisters", "mothers"]
socialism_keywords = ["cooperation", "service", "responsibility", "progress", "social", "grant", "government"]
```

```python
def count_theme(keywords):
    return sum(word_counts[word] for word in keywords if word in word_counts)
nationalism_score = count_theme(nationalism_keywords)
feminism_score = count_theme(feminism_keywords)
socialism_score = count_theme(socialism_keywords)
```

```python
print("Nationalism Score:", nationalism_score)
print("Feminism Score:", feminism_score)
print("Socialism Score:", socialism_score)
```

```python
positive_words = {"happy","inspiring","stirring","privileges","destiny","sacrifice","service", "renaissance","liberal","cooperation","dedication","strengthen", "responsibility","unity","dignity","ideal","brave","strong","good"}
negative_words = {"wrong","stain","reproach","absence"}
pos_count = sum(1 for w in words if w in positive_words)
neg_count = sum(1 for w in words if w in negative_words)
print("Sentiment Analysis")
print("Positive words:", pos_count)
print("Negative words:", neg_count)
plt.figure()
plt.bar(["Positive","Negative"], [pos_count, neg_count])
plt.title("Sentiment Polarity")
plt.ylabel("Frequency")
plt.show()
```

```python
rhetoric_terms = {"duty","destiny","service","responsibility","consecration","dedication"}
rhetoric_counts = {term: words.count(term) for term in rhetoric_terms}
print("\nRhetorical Device Frequency")
for k,v in rhetoric_counts.items():
    print(k, ":", v)
plt.figure()
plt.bar(list(rhetoric_counts.keys()), list(rhetoric_counts.values()))
plt.title("Rhetorical Framing Terms")
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.show()
```

```python
collective = {"we","our","us"}
directive = {"you","your"}
collective_count = sum(1 for w in words if w in collective)
directive_count = sum(1 for w in words if w in directive)
print("\nPronoun Analysis")
print("Collective (we/our/us):", collective_count)
print("Directive (you/your):", directive_count)
plt.figure()
plt.bar(["Collective","Directive"], [collective_count, directive_count])
plt.title("Pronoun Usage")
plt.ylabel("Frequency")
plt.show()
```
